```python
from tensorflow.keras.layers import Embedding ## to implement word2vec
from tensorflow.keras.preprocessing.sequence import pad_sequences  # pad or post-- it can be pre or post padding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size=5000
```

## Onehot Representation

```python
messages = X.copy()
```

```python
messages["title"][1]
```

```python
messages
```

```python
messages.reset_index(inplace=True) # not important 
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
# before one hot encoding we have to preprocessing data
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ## Stemming purpose
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]', ' ',messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[1]
```

```python
onehot_repr=[one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

```python
"""take this one hot representation and apply embedding layer before this,you have check the length of words in doc,not same so we should do the same lenght in order to do this we apply padding"""
```

## Embedding Representation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding="post",maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## Creating model   /----features representation
embedding_vector_features=40   # each and every word is going to get converted by a vector of 40 size (40 features it may have)diferent values how this give 40 features something called crow and skipgram in word2vec.            
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

# Model Training

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

### Adding Dropout=== is basically to improve the performance of the model. it can also reduce overfitting

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0) ##AUC ROC Curve--to find out what should be the probability that you should select
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
